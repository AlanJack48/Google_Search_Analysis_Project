# GOOGLE SEARCH ANALYSIS 

**Pytrends** is an unofficial Google Trends API that provides different methods to download reports of trending results from google trends.

```python
pip install pytrends
```

```python
import pandas as pd
from pytrends.request import TrendReq #Connect to Google
import matplotlib.pyplot as plt
Tr = TrendReq()
```

```python
Tr.build_payload(kw_list=["Machine Learning"]) #Build Payload
data_google = Tr.interest_by_region()
data_google = data_google.sort_values(by="Machine Learning", ascending=False)
data_google = data_google.head(10)
print(data_google)
```

```python
data_google.reset_index().plot(x="geoName", y="Machine Learning", figsize=(15,10), kind="bar")
plt.style.use('dark_background')
plt.show()
```

```python
data_google = TrendReq(hl='en-US', tz=360)
data_google.build_payload(kw_list=['Machine Learning'])
data_google = data_google.interest_over_time() # returns historical, indexed data for when the keyword was searched most as shown on Google Trends.
fig, ax = plt.subplots(figsize=(15, 10))
data_google ['Machine Learning'].plot()
plt.style.use('ggplot')
plt.title('Total Google Searches for Machine Learning', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
